# Particle distributions

```python
#prepare imports
import numpy as np
import astropy.units as u
import naima
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['lines.linewidth'] = 2
```

```python
def particle_distribution(e, f_e):
    return e, (e**2 * f_e).to('erg')
```

naima has the following particle distribution function availablee:

Functional models:
- ``naima.models.PowerLaw``
- ``naima.models.ExponentialCutoffPowerLaw``
- ``naima.models.BrokenPowerLaw``
- ``naima.models.ExponentialCutoffBrokenPowerLaw``
- ``naima.models.LogParabola``

It is necessary to express the amplitude in particles per unit energy.

```python
# energy bounds CTA simulations
e_edges_min = 0.1  # Minimum energy (u.TeV)
e_edges_max = 100. # Maximum energy (u.TeV)
energy_bounds_CTA = [e_edges_min, e_edges_max] * u.TeV
```

```python
# Defines the particles energy 
particles_energy = np.logspace(-5, 5, 1000) * u.TeV
```

### PowerLaw

```python
# naima.models.PowerLaw?
```

```python
PL = naima.models.PowerLaw(
    amplitude=1e36/u.eV, 
    e_0=1*u.TeV, 
    alpha=2.0,
)
```

```python
#check the shape of the particle distribution
plt.loglog(particle_distribution(particles_energy, PL(particles_energy)))
plt.axvline(e_edges_min, ls=':', lw=1, c='k'); plt.axvline(e_edges_max, ls=':', lw=1, c='k')
plt.gca().set_ylim(bottom=1e36, top=1e51)
plt.xlabel('Particle energy [TeV]')
plt.ylabel('$E^2 N(E)$ [erg]')
```

### ExponentialCutoffPowerLaw

```python
ECPL = naima.models.ExponentialCutoffPowerLaw(    
    amplitude=1e36/u.eV, 
    e_0=1*u.TeV, 
    alpha=2.,
    e_cutoff=50*u.TeV,
    beta=.09
)
```

```python
#check the shape of the particle distribution
plt.loglog(particles_energy, particle_distribution(particles_energy, ECPL(particles_energy)))
plt.axvline(e_edges_min, ls=':', lw=1, c='k'); plt.axvline(e_edges_max, ls=':', lw=1, c='k')
plt.gca().set_ylim(bottom=1e36, top=1e51)
plt.xlabel('Particle energy [TeV]')
plt.ylabel('$E^2 N(E)$ [erg]')
```

### BrokenPowerLaw

```python
BPL = naima.models.BrokenPowerLaw(    
    amplitude=1e36/u.eV, 
    e_0=1*u.TeV,  
    e_break=0.2*u.TeV, 
    alpha_1=2.0,
    alpha_2=3.0, 
)
```

```python
#check the shape of the particle distribution
plt.loglog(particles_energy, particle_distribution(particles_energy, BPL(particles_energy)))
plt.axvline(e_edges_min, ls=':', lw=1, c='k'); plt.axvline(e_edges_max, ls=':', lw=1, c='k')
plt.gca().set_ylim(bottom=1e36, top=1e51)
plt.xlabel('Particle energy [TeV]')
plt.ylabel('$E^2 N(E)$ [erg]')
```

### ExponentialCutoffBrokenPowerLaw

```python
ECPL = naima.models.ExponentialCutoffBrokenPowerLaw(
    amplitude=1e36/u.eV, 
    e_0=1*u.TeV, 
    alpha_1=2.0,
    alpha_2=3.0, 
    e_break=0.2*u.TeV, 
    e_cutoff=50*u.TeV,
    beta=2
)
```

```python
#check the shape of the particle distribution
plt.loglog(particles_energy, particle_distribution(particles_energy, ECPL(particles_energy)))
plt.axvline(e_edges_min, ls=':', lw=1, c='k'); plt.axvline(e_edges_max, ls=':', lw=1, c='k')
plt.gca().set_ylim(bottom=1e36, top=1e51)
plt.xlabel('Particle energy [TeV]')
plt.ylabel('$E^2 N(E)$ [erg]')
```
